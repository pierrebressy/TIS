## Function `computeHisto`

Prototype: `def computeHisto(_f)`

This function compututhe the histogram of the given image `_f` and returns the results as an 1D array.
> - `_f` is a 256 gray level impage
> - `h` is the computed histogram

Activities for the `computeHisto` function:
- create the `h` array to receive histogram,
- compute the size of the image `_f`: `height` and `width`,
- go through the while image and construct `h`.

```python
def computeHisto(_f):
    
    # optimized version
    # h,bins=np.histogram(f,bins=np.arange(1+_numBins))

    # naive version
    h=np.zeros( NUM_GRAYS , dtype='uint32')
    height=_f.shape[0]
    width=_f.shape[1]
    for l in range(0,height):
        for c in range(0,width):
            pixel=_f[l][c];
            h[pixel]=h[pixel]+1
    return h;
```

## Test of the `computeHisto` function

This cell creates an array with random values (256x256) and computes:
- the histogram using `numpy.histogram` (=> `theoreticalResult`),
- the histogram usin the `computeHisto` function (=> `practicalResult`).

It compares the two results using the `np.testing.assert_equal` function which generate an error if the test fails.

```python
# create the test array
testArraySize=10
testArray=(np.random.rand(testArraySize,testArraySize)*NUM_GRAYS).astype('uint8')
df=pd.DataFrame(testArray)
print("test array\n",df)

# compute theoretical histogram 
theoreticalResult,bins=np.histogram(testArray,bins=np.arange(1+NUM_GRAYS))
#print("theoreticalResult",theoreticalResult)

# compute our histogram 
practicalResult=computeHisto(testArray);
#print("practicalResult",practicalResult)

# compare results 
np.testing.assert_equal(theoreticalResult, practicalResult)
print("TEST computeHisto OK")
```

## Function to compute the cumulative histogram 

Prototype: `def computeCumulativeHisto(_h)`

This function computes the cumulative histogram of the histogram of an image and returns the results and the normalized results (0..255) as two 1D arrays.
> - `_h` is the histogram
> - `cumulativeHisto` is the cumulative histogram
> - `normalizedCumulativeHisto` is the normalized cumulative histogram (0..255)

**Activities for the `computeCumulativeHisto` function:**
- create the `cumulativeHisto` and `normalizedCumulativeHisto` arrays to receive the cumulative histograms,
- compute the cumulative histogram `cumulativeHisto`,
- normalize the cumulative histogram `cumulativeHisto` to `normalizedCumulativeHisto`.

```python
def computeCumulativeHisto(_h):

    # optimized version
    # cumulativeHisto=np.cumsum(_h)
    # normalizedCumulativeHisto=(MAX_GRAY*cumulativeHisto/cumulativeHisto[-1]).astype('uint8')
    
    # naive version
    cumulativeHisto=np.zeros( (NUM_GRAYS) , dtype='uint32')
    normalizedCumulativeHisto=np.zeros( (NUM_GRAYS) , dtype='uint32')
    cumulativeHisto[0]=_h[0]
    for k in range(1,NUM_GRAYS):
        cumulativeHisto[k]=cumulativeHisto[k-1]+_h[k]

    maxVal=cumulativeHisto[(NUM_GRAYS-1)]
    for k in range(0,NUM_GRAYS):
        normalizedCumulativeHisto[k]=MAX_GRAY*cumulativeHisto[k]/maxVal.astype('double')
    
    return cumulativeHisto, normalizedCumulativeHisto;
```

## Test of the `computeCumulativeHisto` function

This cell creates an array with random values (256x256) and computes:
- the histogram using `numpy.histogram` 
- the cumulative histogram using `numpy.cumsum` (=> `theoreticalResult`),
- the histogram usin the `computeHisto` function,
- the cumulative histogram using the `computeCumulativeHisto`function (=> `practicalResult`).

It compares the two results using the `np.testing.assert_equal` function which generate an error if the test fails.

```python
# create the test array
testArraySize=10
testArray=(np.random.rand(testArraySize,testArraySize)*NUM_GRAYS).astype('uint8')
df=pd.DataFrame(testArray)
print("test array\n",df)

# compute theoretical histogram 
theoreticalHisto,bins=np.histogram(testArray,bins=np.arange(1+NUM_GRAYS))
theoreticalCumulativeHisto=np.cumsum(theoreticalHisto)
theoreticalNormalizedCumulativeHisto=(MAX_GRAY*theoreticalCumulativeHisto/theoreticalCumulativeHisto[-1]).astype('uint8')
#print("theoreticalCumulativeHisto",theoreticalCumulativeHisto)
#print("theoreticalNormalizedCumulativeHisto",theoreticalNormalizedCumulativeHisto)

# compute our histogram 
practicalHisto=computeHisto(testArray);
practicalCumulativeHisto, practicalNormalizedCumulativeHisto=computeCumulativeHisto(practicalHisto)
#print("practicalCumulativeHisto",practicalCumulativeHisto)
#print("practicalNormalizedCumulativeHisto",practicalNormalizedCumulativeHisto)

# compare results 
np.testing.assert_equal(theoreticalCumulativeHisto, practicalCumulativeHisto)
np.testing.assert_equal(theoreticalNormalizedCumulativeHisto, practicalNormalizedCumulativeHisto)
print("TEST computeCumulativeHisto OK")
```

## Function to apply a Look Up Table (LUT) to an image 

Prototype: `def applyLUT(_f, _lut)`

This function applies the _lut to _f and returns the results as 2D array.
> - `_f` is the image to apply the LUT,
> - `_lut` is the LUT,
> - `g` is the output.

**Activities for the `applyLUT` function:**
- create the `g` array to receive the output image,
- compute `g`.


```python
def applyLUT(_f,_lut):
    
    # optimized version
    # g=_lut[_f]
    
    # naive version
    height=_f.shape[0]
    width=_f.shape[1]
    g=np.zeros( (height,width), dtype='uint8')
    for l in range(0,height):
        for c in range(0,width):
            pixel=_f[l][c];
            g[l][c]=_lut[pixel]
    g=_lut[_f]
    return g;
```

## Function to adjust the histogram between 1 and 99% 

Prototype: `def imgLevelAdjust(_f)`

This function applies the _lut to _f and returns the results as 2D array.
> - `_f` is the image to adjust the histogram,
> - `h` is the output.

**Activities for the `imgLevelAdjust` function:**
- compute the histogram of `_f`,
- find the 2 limits (dark and white) in the cumulative histogram,
- create the LUT,
- compute `h` by applying the LUT to `_f`.


```python
def imgLevelAdjust(_f):

    histo=computeHisto(_f)
    cumulativeHisto,normalizedCumulativeHisto=computeCumulativeHisto(histo)
    dark_sum_limit = cumulativeHisto[(NUM_GRAYS-1)] * 1 / 100;
    white_sum_limit = cumulativeHisto[(NUM_GRAYS-1)] * 99 / 100;

    index_dark_sum_limit=0;
    while(cumulativeHisto[index_dark_sum_limit]<dark_sum_limit):
        index_dark_sum_limit+=1

    index_white_sum_limit=MAX_GRAY;
    while(cumulativeHisto[index_white_sum_limit]>white_sum_limit):
        index_white_sum_limit-=1

    lut=np.zeros( (NUM_GRAYS), dtype='uint8')
    lut[0:index_dark_sum_limit]=0
    lut[index_white_sum_limit:(NUM_GRAYS-1)]=MAX_GRAY
    for k in range(index_dark_sum_limit,index_white_sum_limit):
        lut[k]=(k-index_dark_sum_limit) * MAX_GRAY /(index_white_sum_limit-index_dark_sum_limit)

    h=applyLUT(f,lut)    
    return h
 
```

## Ex 3: Halftoning

```python
def halfToning(_f):
    height=_f.shape[0]
    width=_f.shape[1]
    ht=np.zeros( (height,width) , dtype='uint8')
    _h=np.zeros( (height,width) , dtype='double')

    for l in range(0,height):
        for c in range(1,width):
            _h[l][c]=_f[l][c];

    for l in range(0,height-1):
        for c in range(1,width-1):

            if _h[l][c]>128:
                ht[l][c]=255
            else:
                ht[l][c]=0

            err=_h[l][c]-ht[l][c];
            _h[l  ][c+1]+=7/16*err 
            _h[l+1][c-1]+=3/16*err
            _h[l+1][c ]+=5/16*err
            _h[l+1][c+1]+=1/16*err

    return ht 
 
```

```python
ht=halfToning(f)
plt.rcParams['figure.figsize']=[10,10] #inches !
plt.imshow(ht, cmap='gray')
sk.io.imsave('data/output3.png',ht)
```

```python
# transform image to C 2D array
n='pout'
f=sk.io.imread('data/'+n+'.bmp', True) # True => convert to gray
f=255*np.array(f)
f=f.astype('uint8')
height=f.shape[0]
width=f.shape[1]


fd = open('data/'+n+'.h','w') 
fd.write('#ifndef _'+n+'_H_\n');
fd.write('#define _'+n+'_H_\n\n');
fd.write('#include <stdint.h>\n\n');
fd.write('#define IMG_HEIGHT ');
fd.write(str(height));
fd.write('\n');
fd.write('#define IMG_WIDTH ');
fd.write(str(width));
fd.write('\n\n');
fd.write('extern const uint8_t '+n+'[IMG_HEIGHT][IMG_WIDTH];\n\n');
fd.write('#endif // '+n+'\n');
fd.close()

fd = open('data/'+n+'.c','w') 
fd.write('#include "'+n+'.h"\n\n');

fd.write('const uint8_t '+n+'[IMG_HEIGHT][IMG_WIDTH]={\n');
           
       
for l in range(0,height):
    fd.write('{ ')
    for c in range(0,width):
        fd.write(str(f[l][c]))
        if(c<width-1):
            fd.write(',')
                
    if(l<height-1):
        fd.write('},\n')
    else:
        fd.write('}\n')
fd.write('};\n');

fd.close()
```
